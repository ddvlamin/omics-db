```python
import json

from pymilvus_orm import connections, FieldSchema, CollectionSchema, DataType, Collection
import h5py
import pymilvus_orm
import numpy as np
```

```python
files = ["../data/pdbaa-1.h5","../data/pdbaa-2.h5"]
```

```python
def save_vector_file(fileindex, batch):
    vector_file = f"../../data/npy/vectors/index2id_{fileindex}.npy"
    
    with open(vector_file, "wb") as fout:
        np.save(fout, batch) 
    print(f"saved file {vector_file} with {len(batch)} vectors")  
    
block_size = 1000
dim = 6165
batch = np.zeros((block_size, dim))
meta = dict()
fileindex = 1
filesizes = [0]*(len(files)+1)

for i, file in enumerate(files):
    f = h5py.File(file, 'r')
    for j, key in enumerate(f.keys()):
        filesizes[i+1] += 1
        batch[j%block_size] = np.array(f[key])
        meta[filesizes[i]+j] = key

        if (j+1)%block_size == 0:
            save_vector_file(fileindex, batch)
            fileindex += 1
            batch = np.zeros((block_size, dim))
            
    if (j+1)%block_size != 0:
        save_vector_file(fileindex, batch[0:(j+1)%block_size])
        fileindex += 1
        batch = np.zeros((block_size, dim))
        
    f.close()
    
meta_file = f"../../data/npy/meta/index2id.json"
with open(meta_file, "w") as fout:
    json.dump(meta, fout)    
print(f"saved file {meta_file}")  
```
